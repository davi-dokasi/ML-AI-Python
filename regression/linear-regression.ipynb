```python
# Core Pkg's 

import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
dt = pd.read_csv("data/Salary_Data.csv")

dt
```

```python
X = dt.iloc[:, :-1].values
y = dt.iloc[:, -1].values

X,y
```

```python
plt.scatter(X, y, color = 'red')

plt.title("Salário vs Experiência")
plt.xlabel("Experiência"), plt.ylabel("Salário")

plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x_train, y_train) # ajusta o modelo de regressão linear aos dados de treino
```

#### Predicting the Test set results:

```python
y_pred = reg.predict(x_test) # prevê os valores de y para os dados de teste
```

```python
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, reg.predict(x_train), color = 'blue')

plt.title('Salário vs Experiência (Conjunto de Treino)')
plt.xlabel('Anos de Experiência'), plt.ylabel('Salário')
plt.show()
```

#### Visualising the Test set results

```python
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, reg.predict(x_train), color = 'blue')

plt.title('Salário vs Experiência (Conjunto de Teste)')
plt.xlabel('Anos de Experiência'), plt.ylabel('Salário')
plt.show()
```

#### Exemple of a single prediction

```python
print(reg.predict([[12]]))
```

Therefore, our model predicts that the salary of an employee with 12 years of experience is $ 138967,5.

**Important note:** Notice that the value of the feature (12 years) was input in a double pair of square brackets. That's because the "predict" method always expects a 2D array as the format of its inputs. And putting 12 into a double pair of square brackets makes the input exactly a 2D array. Simply put:

$12 \rightarrow \textrm{scalar}$

$[12] \rightarrow \textrm{1D array}$

$[[12]] \rightarrow \textrm{2D array}$

#### The coefficients

```python
# print(reg.coef_)
# print(reg.intercept_)

reg.coef_, reg.intercept_
```

Therefore, the equation of our simple linear regression model is:

$$\textrm{Salary} = 9345.94 \times \textrm{YearsExperience} + 26816.19$$

**Important Note:** To get these coefficients we called the "coef_" and "intercept_" attributes from our regressor object. Attributes in Python are different than methods and usually return a simple value or an array of values.
